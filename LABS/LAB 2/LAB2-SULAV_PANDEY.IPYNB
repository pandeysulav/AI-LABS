Reading & Cleaning CSV Files

```python
# 1. Read a CSV file using pandas read_csv() and display its contents.

df= pd.read_csv('Iris.csv')
df = df.drop(columns=['Id'])
df
```

```python
# 2. Print the column names, number of rows, and number of columns.

df.info()
```

```python
# 3. Display the first 5 rows and check for missing values.

print(df.head(5))
print(df.isnull().sum())
```

```python
# 4. Drop rows that contain any missing values using dropna().

df_cleaned = df.dropna()
df_cleaned
```

```python
# 5. Fill missing numeric values with the column mean using fillna()
numeric_cols = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())

# verify there are no missing values left
print(df.isnull().sum())
```

```python
# 6. Remove duplicate rows from the dataset.
print(df.duplicated().sum())
df = df.drop_duplicates()

# optional: reset the index after dropping rows
df = df.reset_index(drop=True)

print(df.shape)
```

```python
# 7. Rename column headers to lowercase using str.lower().

df.columns = df.columns.str.lower()
print(df.columns)
```

```python
# 8. Convert a categorical column (e.g. Gender: Male/Female) into numbers using map() or replace().

df["species"] = df["species"].replace({
    "Iris-setosa": 0,
    "Iris-versicolor": 1,
    "Iris-virginica": 2
})
df
```

```python
# 9. Select only numeric columns and print their correlation using corr().

print(df.select_dtypes(include=['number']).corr())
```

```python
# 10. Save the cleaned DataFrame to a new CSV file using to_csv().

df.to_csv('cleaned_iris.csv', index=False)

print("DataFrame successfully saved as 'cleaned_iris.csv'")
```
